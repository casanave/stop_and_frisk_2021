### Precinct and Index as DateTime

```python
data['STOP_LOCATION_PRECINCT'] = data['STOP_LOCATION_PRECINCT'].astype(str)
data.set_index(data['STOP_FRISK_DATE_STOP_FRISK_TIME'], inplace = True)
data.drop(columns = 'STOP_FRISK_DATE_STOP_FRISK_TIME', inplace = True)
data.head()

# casting precinct to a string so no calculations are used on that column
# making the date time the index
```

### Suspect Reported Age 

```python
for x in data['SUSPECT_REPORTED_AGE'].value_counts().index:
    print (x)
```

```python
data['SUSPECT_REPORTED_AGE'].replace({0 : "UNKNOWN", 120 : "UNKNOWN"}, 
                                     inplace = True)

# reassigning "unknown" for illogical ages
```

```python
age_known = data.loc[data['SUSPECT_REPORTED_AGE'] != 'UNKNOWN']

age_known['SUSPECT_REPORTED_AGE'] = age_known['SUSPECT_REPORTED_AGE'].astype(int)

# making a new df just of rows where the reported age is known
```

```python
for x in [0.25, 0.50, 0.75]:
    print(age_known['SUSPECT_REPORTED_AGE'].quantile(x))
    
# finding the quantiles of suspect reported age so I can make age bins 
```

```python
bins = [ 0, 20, 28, 38, 100]


labels = ['Twenty and Under', 'Twentyone-Twentyeight', 'Twentyeight-Thirtyeight',
          'Thirtyeight-Above']

cut_data, bin_edges = pd.cut(age_known['SUSPECT_REPORTED_AGE'], 
                             bins = bins,
                             labels = labels, retbins = True)
bin_edges

# making categorical bins for age categories 
```

```python
age_known['AGE_CATEGORY'] = cut_data
print(age_known['SUSPECT_REPORTED_AGE'], age_known['AGE_CATEGORY'])


# making the new columns and taking a look at the data 
```

```python
age_unknown = data.loc[data['SUSPECT_REPORTED_AGE'] == 'UNKNOWN']
age_unknown['AGE_CATEGORY'] = 'UNKNOWN'
print(age_unknown['SUSPECT_REPORTED_AGE'], age_unknown['AGE_CATEGORY'])

# making unknown age rows their own df and making it have an age category column 
# then taking a look at that new df
```

```python
dfs = [age_known, age_unknown]

data = pd.concat(dfs)
data.sort_index(inplace = True)
data.head()

# putting the two dfs together and resorting them by date time 
```

```python
data['AGE_CATEGORY'].value_counts()

# taking a look at the distributions of the age category column 
```

### Observations: 

Suspect Arrested and Suspect Not Arrested:

- The overall average ratio is 37% Arrested to 62% Not Arrested on Average with a standard deviation of about 50% which means that the variability is higher than the actual arrest rate. 

- The average ratio of firearm being drawn during stops is just under 6% of the time. 

- Handcuffs are being enforced about 20% of the time during stops. 

- OC spray wasn't used during 2021 during stops. 

- Physical force "other" was used about 3% of the time during stops. 

- Restraint was used just under 3% of the time during stops. 

- Physical Force was used 90% of the time, making it an regular part of most stops. 

- Weapon Impact was hardly ever recorded as being used. 

- The average reported age was 27 years, and I'll explore this more in visualization. 

- The physical force total, perhaps the most interesting metric in this set of stats is at about 1.25% meaning if we sum up all the various counts of physical force used, there were more counts of physical force used than there were stops by about a quarter. This is a very interesting statistic. With the standard deviation being at about 0.5 or a half count of a kind of physical force used, that means that the average experience was that physical force was used on a stop, and how much physical force normally varied by a half count.
